```python
import pandas as pd
```

```python
df = pd.read_csv("./My Data/data.csv")
```

```python
df = df.drop(list(df)[0], axis = 1)
```

```python
df.head()
```

```python
def generateTargets(df):
    df["Target"] = df["Market Cap"].shift(-1)
    return df[:-1]
```

```python
df = df.dropna(subset=["Company", "Date"])
```

```python
df = df.sort_values(["Company", "Date"])
```

```python
df1 = df.groupby("Company").apply(
    lambda group: generateTargets(group)
)
```

```python
df1 = df1.droplevel(0)
```

```python
df1.head()
```

```python
df1 = df1.dropna()
```

```python
df1.to_csv("./My Data/final_cleaned_data.csv")
```

```python
df1["Test"] = 100*(df1["Target"] - df1["Market Cap"])/df1["Market Cap"]
```

```python
df1['Label'] = 1
df1.loc[df1['Test'] > 5, 'Label'] = 2
df1.loc[df1['Test'] < -5, 'Label'] = 0

# Label = 2 indicates buy
# Label = 1 indicates hold
# Label = 0 indicates sell
```

```python
df1["Target"].head()
```

```python
df1["Market Cap"].head()
```

```python
df1.head()
```

```python
df1 = df1.drop("Test", axis=1)
```

```python
normalized = pd.DataFrame()
norm_ratios = dict()
```

```python
for k in list(df1.keys()):
    if k != "Company" and k != "Date" and k != "Label":
        normalized[k] = (df1[k] - df1[k].mean()) / df1[k].std()
        norm_ratios[k] = df1[k].mean()
    else:
        normalized[k] = df1[k]
```

```python
normalized
```

```python
pd.DataFrame(norm_ratios, index=[0]).to_csv("./My Data/norm_ratios.csv", index=False)
normalized.reset_index(drop=True).to_csv("./My Data/normalized_data.csv", index=False)
```

```python
test = normalized[normalized["Date"] >= "2018"].reset_index(drop=True)
```

```python
other = normalized[normalized["Date"] <= "2018"].reset_index(drop=True)
other = other.sample(frac=1).reset_index(drop=True)
```

```python
length = len(other)

train_ratio = 0.95
train_samples = (int)(length*train_ratio)
```

```python
train = other[0:train_samples]
val = other[train_samples:]
```

```python
train.to_csv("./My Data/train.csv", index=False)
val.to_csv("./My Data/val.csv", index=False)
test.to_csv("./My Data/test.csv", index=False)
```
